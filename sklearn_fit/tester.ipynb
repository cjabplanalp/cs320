```python
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
users = pd.read_csv('data/train_users.csv')
y = pd.read_csv('data/train_y.csv')
logs = pd.read_csv('data/train_logs.csv')

test_users = pd.read_csv("data/test1_users.csv")
test_logs = pd.read_csv("data/test1_logs.csv")
test_y = pd.read_csv('data/test1_y.csv')
```

```python
df = users.merge(y)
df
```

```python
model = LogisticRegression()
model.fit(df[['past_purchase_amt']], df[['y']])
model.score(df[['past_purchase_amt']], df[['y']])
```

```python
edit = logs.groupby('user_id', as_index = False).sum()
df = df.merge(edit, on = 'user_id', how = 'left')
```
